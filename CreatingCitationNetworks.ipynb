```python
import pandas as pd

df_authors = pd.read_csv("medicine_papers_with_authors.csv")

df_refs = pd.read_csv("PaperReferences.csv")

df_gender = pd.read_csv("AuthorsGenderized.csv")[["AuthorId", "Gender"]]

paper_ids = df_authors["PaperId"].unique()
df_refs_filtered = df_refs[
    df_refs["PaperId"].isin(paper_ids) &
    df_refs["PaperReferencedId"].isin(paper_ids)
]

df_citing = df_refs_filtered.merge(df_authors, on="PaperId")
df_citing = df_citing.rename(columns={"AuthorId": "CitingAuthorId"})

df_cited = df_refs_filtered.merge(df_authors, left_on="PaperReferencedId", right_on="PaperId")
df_cited = df_cited.rename(columns={"AuthorId": "CitedAuthorId"})

df_edges = df_citing[["PaperId", "PaperReferencedId", "CitingAuthorId"]] \
    .merge(df_cited[["PaperReferencedId", "CitedAuthorId"]], on="PaperReferencedId")

df_edges = df_edges \
    .merge(df_gender, left_on="CitingAuthorId", right_on="AuthorId", how="left") \
    .rename(columns={"Gender": "CitingGender"}) \
    .drop(columns=["AuthorId"])

df_edges = df_edges \
    .merge(df_gender, left_on="CitedAuthorId", right_on="AuthorId", how="left") \
    .rename(columns={"Gender": "CitedGender"}) \
    .drop(columns=["AuthorId"])

df_edges = df_edges[["CitingAuthorId", "CitingGender", "CitedAuthorId", "CitedGender"]].drop_duplicates()
df_edges.to_csv("medicine_author_citation_gendered.csv", index=False)

print("Saved citation network with gender info.")
```
