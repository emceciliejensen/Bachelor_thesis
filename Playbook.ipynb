```python
import pandas as pd
#Getting field of study id's for medicine, humanities, engineering, psychology, physics.

df = pd.read_csv("field_of_study.csv")
#medicine
medicine_FOS_id = df[df["NormalizedName"] == "medicine"]["FieldOfStudyId"].values
print("medicine_FOS_id=",medicine_FOS_id)
#humanities
humanities_FOS_id = df[df["NormalizedName"] == "humanities"]["FieldOfStudyId"].values
print("humanities_FOS_id=",humanities_FOS_id)
#engineering
engineering_FOS_id = df[df["NormalizedName"] == "engineering"]["FieldOfStudyId"].values
print("engineering_FOS_id",engineering_FOS_id)
#psychology
psychology_FOS_id = df[df["NormalizedName"] == "psychology"]["FieldOfStudyId"].values
print("psychology_FOS_id",psychology_FOS_id)
#physics
physics_FOS_id = df[df["NormalizedName"] == "physics"]["FieldOfStudyId"].values
print("physics_FOS_id",physics_FOS_id)
```
